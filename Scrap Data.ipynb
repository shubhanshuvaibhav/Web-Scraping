# Popular Variants of Oats


The focus of my project was to identify the most popular variants of oats on Flipkart, based on their ratings and the number of people who rated them. To begin with, I extracted the required data by scraping information from a website (www.flipkart.com) and utilized the pandas library for further analysis.

```python
#Importing libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Flipkart oats search link and user agent
URL = 'https://www.flipkart.com/search?q=oats+1+kg&sid=eat%2Ccpy&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=oats+1+kg%7CBreakfast+Cereal&requestId=f7042106-a034-4cbb-8c1a-1238914361a0&as-searchtext=oats'
Headers = ({"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 Edg/121.0.0.0","Accept-Language":"en-US, en;q=0.5"})
```

```python
#sending requests to the url

webpage = requests.get(URL, headers = Headers)
```

```python
#creating soup object
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
#finding all the links
links1 = soup.find_all("a",attrs={'class':'s1Q9rs'})
```

```python
links_list = []
```

```python
#looping through object for extracting links

for link in links1:
    links_list.append(link.get('href'))
```

```python
#scraping data and adding it to dictionary

d = {"title":[], "price":[], "rating":[], "num_of_people_rated":[]}

for link in links_list:
    new_webpage = requests.get("https://www.flipkart.com" + link, headers=Headers)
    
    new_soup = BeautifulSoup(new_webpage.content,'html.parser')
    
    t = new_soup.find('span',{'class':'B_NuCI'})
    d['title'].append(t.text)
    
    p = new_soup.find('div',{'class':'_30jeq3 _16Jk6d'})
    d['price'].append(p.text)
        
    r = new_soup.find('div',{'class':'_3LWZlK'})
    d['rating'].append(r.text)
    
    n = new_soup.find('span',{'class':"_2_R_DZ"})
    d['num_of_people_rated'].append(n.text)
```

```python
# cleaning data for num_of_people_rated

Reviews =[]

for r in d['num_of_people_rated']:
    parts = (r.split('&'))[0].split(" ")
   
    
    Reviews.append(parts[0])

Reviews
```

```python
# adding it to dictionary
d['num_of_people_rated'] = Reviews
```

```python
# removing ',' from reviews and '₹' from price

a=[]
for r in d['num_of_people_rated']:
    b = r.replace(',','')
    a.append(b)
d['num_of_people_rated'] = a

pr=[]
for p in d['price']:
    i = (p.replace('₹','')).replace(',','')
    pr.append(i)
d['price'] = pr
```

```python
#coverting to dataframe and storing it in csv file
oat_df = pd.DataFrame.from_dict(d)
oat_df.to_csv("oats_data.csv")
```

```python
oat_df.dtypes
```

```python
#changing data type
oat_df['price'] = oat_df['price'].astype('int')
oat_df['rating'] = oat_df['rating'].astype('float')
oat_df['num_of_people_rated'] = oat_df['num_of_people_rated'].astype('int')
```

```python
# sorted dataframe based on poularity and rating

oat_df.sort_values(by=['num_of_people_rated','rating'], ascending = False, inplace = True)
```
